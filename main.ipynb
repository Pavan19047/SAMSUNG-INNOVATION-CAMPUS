```python
def max_pooling(matrix, pool_size):
    pooled_matrix = []
    for i in range(0, matrix.shape[0], pool_size):
        row = []
        for j in range(0, matrix.shape[1], pool_size):
            pool = matrix[i:i+pool_size, j:j+pool_size]
            row.append(np.max(pool))
        pooled_matrix.append(row)
    return np.array(pooled_matrix)
```

```python
max_pooled_matrix = max_pooling(random_matrix, 2)
plt.imshow(max_pooled_matrix, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.show()
```

```python
import cv2

# Open the video file
cap = cv2.VideoCapture("video.mp4")

# Read all frames into a list
frames = []
while True:
    ret, frame = cap.read()
    if not ret:
        break
    frames.append(frame)

cap.release()

# Reverse the list of frames
frames.reverse()

# Get video properties for saving
height, width, layers = frames[0].shape
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('reversed_video.mp4', fourcc, 30, (width, height))

# Write reversed frames to new video
for frame in frames:
    out.write(frame)

out.release()
print("Reversed video saved as 'reversed_video.mp4'")
```

```python
import pandas as pd

baby_height = [30,32,34,28,38,26,40,44,36,28,40,39,37]
baby_weight = [2.8,3,2.7,3.6,2.2,3.8,4.1,3.1,2.8,2.9,3.6,3.9,2.8]

df = pd.DataFrame({
    'Height': baby_height,  
    'Weight': baby_weight
})


df['Predicted_Weight'] = 0.1 * df['Height'] - 0.05

print(df)
```

```python
df['Squared_Error'] = (df['Weight'] - df['Predicted_Weight']) ** 2

df['Mean_squared_Error'] = df['Squared_Error'].mean()
print(df)
```

```python
df["Gender"] = ["F","F","F","M","F","M","M","F","F","M","F","F","F"]
df
```

```python
df['Predicted_Gender'] = df.apply(lambda row: 'M' if row['Weight'] > 0.1 * row['Height'] - 0.05 else 'F', axis=1)
plt.scatter(df['Height'], df['Weight'], c=df['Predicted_Gender'].map({'M': 'blue', 'F': 'magenta'}), label='Predicted Gender')

x_vals = df['Height']
y_vals = 0.1 * x_vals - 0.05
plt.plot(x_vals, y_vals, color='green', linestyle='--', label='Classification Line')

plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Predicted Gender based on Height and Weight')
plt.legend()
plt.show()
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Height'], df['Weight'], color='blue', label='Actual Data')
plt.plot(df['Height'], df['Predicted_Weight'], color='red', label='Predicted Line')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight with Linear Regression')
plt.legend()
plt.show()
```

```python
total_bill = [10,15,20,25,30,35,40]
tips = [0.2,0.22,0.26,0.29,0.30,0.36,0.38]
df2 = pd.DataFrame({
    'Total_Bill': total_bill,
    'Tips': tips
})
```

```python
tips_predicted = 0.3 * df2['Total_Bill'] + 0.05
df2['Predicted_Tips'] = tips_predicted
df2['Squared_Error'] = (df2['Tips'] - df2['Predicted_Tips']) ** 2
df2['Mean_squared_Error'] = df2['Squared_Error'].mean()
print(df2)
```

```python
#visualize
import matplotlib.pyplot as plt
plt.scatter(df2['Total_Bill'], df2['Tips'], color='blue', label='Actual Data')
plt.plot(df2['Total_Bill'], df2['Predicted_Tips'], color='red', label='Predicted Line')
plt.xlabel('Total Bill')
plt.ylabel('Tips')
plt.title('Total Bill vs Tips with Linear Regression')
plt.legend()
plt.show()
```
